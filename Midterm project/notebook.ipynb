```python
def cargar_dataset_kagglehub(dataset_id):
    import kagglehub, os, pandas as pd

    path = kagglehub.dataset_download(dataset_id)
    for root, dirs, files in os.walk(path):
        for file in files:
            full_path = os.path.join(root, file)
            print("Archivo encontrado:", full_path)
            if file.endswith(".csv"):
                return pd.read_csv(full_path)
            elif file.endswith(".xlsx"):
                return pd.read_excel(full_path)
    raise FileNotFoundError("No se encontró ningún archivo CSV o Excel en el dataset.")

df = cargar_dataset_kagglehub("wafaaelhusseini/wine-and-food-pairing-dataset")

```
